```python
# Importing libarires
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Preparping Algorthims Leaderboard Dataset:

```python
# Information about algorthims problems
algo_info = pd.read_csv("algo_info.csv")
# Select the problems that were only sampled
algo_info = algo_info.iloc[np.arange(0,len(algo_info),1/0.1)]
# Adding new column 'problem_id'
algo_info.insert(0,"problem_id",np.arange(0,len(algo_info)))
algo_info.head()
```

```python
# Algorthims dataset
algo_leaderboard = pd.read_csv("algo_leaderboard.csv", sep=";",error_bad_lines = False)
# Adding new column 'problem_type'
algo_leaderboard.insert(1,"problem_type", "algorithms")
# Change range of of 'problem_id'
algo_leaderboard["problem_id"] = algo_leaderboard["problem_id"]/10
# Casting appropritate datatypes
algo_leaderboard[["problem_id","count"]] = algo_leaderboard[["problem_id","count"]].astype(np.int16)
algo_leaderboard["score"] = algo_leaderboard["score"].astype(np.float16)
algo_leaderboard.head(10)
```

```python
algo_leaderboard["problem_id"].unique()
```

### Preparping Data Structures Leaderboard Dataset:

```python
# Information about algorthims problems
ds_info = pd.read_csv("ds_info.csv")
# Select the problems that were only sampled
ds_info = ds_info.iloc[np.arange(0,len(ds_info),1/0.1)]
# Adding new column 'problem_id'
ds_info.insert(0,"problem_id",np.arange(len(algo_info),len(ds_info)+len(algo_info)))
ds_info.head()
```

```python
# Data structures dataset
ds_leaderboard = pd.read_csv("ds_leaderboard.csv", sep=";",error_bad_lines = False)
# Adding new column 'problem_type'
ds_leaderboard.insert(1,"problem_type", "data structures")
# Change range of of 'problem_id'
ds_leaderboard["problem_id"] = ds_leaderboard["problem_id"]/10 + len(algo_info)
# Casting appropritate datatypes
ds_leaderboard[["problem_id","count"]] = ds_leaderboard[["problem_id","count"]].astype(np.int16)
ds_leaderboard["score"] = ds_leaderboard["score"].astype(np.float16)
ds_leaderboard.head(10)
```

```python
ds_leaderboard["problem_id"].unique()
```

### Outputting Files After Preperation:

```python
# Combined info of algorthims and data structures problems
algo_ds_info = pd.concat([algo_info,ds_info], axis = 0)
algo_ds_info
```

```python
# Combined algorthims and data structures leaderboard datasets
algo_ds_leaderboard = pd.concat([algo_leaderboard,ds_leaderboard], axis = 0)
algo_ds_leaderboard.head()
```

```python
algo_ds_leaderboard.to_csv("algo_ds_leaderboard.csv",sep=";",index=False)
algo_ds_info.to_csv("algo_ds_info.csv",index=False)
```
